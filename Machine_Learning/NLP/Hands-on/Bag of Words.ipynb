```python
import pandas as pd
import numpy as np
import collections
```

```python
doc1 = "Harry Potter is an amazing movie!!, harry"
doc2 = "Harry Potter is the best movie!"
doc3 = "Harry potter is so great"
```

```python
import re
doc1 = re.sub(r"[^A-Za-z0-9]"," ",doc1.lower()).split()
doc2 = re.sub(r"[^A-Za-z0-9]"," ",doc2.lower()).split()
doc3 = re.sub(r"[^A-Za-z0-9]"," ",doc3.lower()).split()
```

```python
unique_words = set(doc1+doc2+doc3)
unique_words
```

```python
def bagofWords(unique_words, text):
    bow = dict.fromkeys(unique_words,0)
    for t in text:
        bow[t] = text.count(t) 
    return bow
```

```python
bow1 = bagofWords(unique_words, doc1)
bow2 = bagofWords(unique_words, doc2)
bow3 = bagofWords(unique_words, doc3)
```

```python
bow3
```

```python
df_bow = pd.DataFrame([bow1,bow2,bow3])
df_bow
```

### Sklearn

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True)
doc1 = "Harry Potter is an amazing movie!!, harry"
doc2 = "Harry Potter is the best movie!"
doc3 = "Harry potter is so great"
cv_out = cv.fit_transform([doc1,doc2,doc3]).toarray()
cv_df = pd.DataFrame(cv_out, columns=cv.get_feature_names_out())
cv_df
```

```python
cv = CountVectorizer()
doc1 = "Harry Potter is an amazing movie!!, harry"
doc2 = "Harry Potter is the best movie!"
doc3 = "Harry potter is so great"
cv_out = cv.fit_transform([doc1,doc2,doc3]).toarray()
cv_df = pd.DataFrame(cv_out, columns=cv.get_feature_names_out())
cv_df
```

```python
cv = CountVectorizer(ngram_range=(1,2)) #bigram, trigram, ngram
doc1 = "Harry Potter is an amazing movie!!, harry"
doc2 = "Harry Potter is the best movie!"
doc3 = "Harry potter is so great"
cv_out = cv.fit_transform([doc1,doc2,doc3]).toarray()
cv_df = pd.DataFrame(cv_out, columns=cv.get_feature_names_out())
cv_df
```
